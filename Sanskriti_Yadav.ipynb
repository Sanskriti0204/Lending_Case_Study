### Cleaning and manipulating data

```python
#checking null values in all the columns
    
#creating function to check for the null values in the dataset
def null_values(dataset):
    return round((dataset.isnull().sum()*100/len(dataset)).sort_values(ascending=False),2)
```

```python
null_values(df)
```

```python
#dealing with null values >50%


null_50 = null_values(df)[null_values(df)>50]

print(null_50)
print("Num of columns having missing values more than 50% :",len(null_50))
```

###### Categorical variable analysis

```python
def categoricalfn(df,col,vartype,hue =None):
    sns.set(style="darkgrid")   
    if vartype == 1:
        temp = pd.Series(data = hue)
        fig, ax = plt.subplots()
        width = len(df[col].unique()) + 6 + 4*len(temp.unique())
        fig.set_size_inches(width , 7)
        ax = sns.countplot(data = df, x= col, order=df[col].value_counts().index,hue = hue) 
        if len(temp.unique()) > 0:
            for p in ax.patches:
                ax.annotate('{:1.1f}%'.format((p.get_height()*100)/float(len(df))), (p.get_x()+0.05, p.get_height()+20))  
        else:
            for p in ax.patches:
                ax.annotate(p.get_height(), (p.get_x()+0.32, p.get_height()+20)) 
        del temp
    else:
        exit


# Plotting graph for loan status
categoricalfn(df,col='loan_status',vartype=1)
df['loan_status'].count()
```

2.87% of the applicants are charged off.

```python
#plotting graph for loan purpose
categoricalfn(df,col='purpose',vartype=1,hue='loan_status')
```

About 47% people took loan to pay off other debts and loans.

# Multivariate Analysis
Checking which parameters are impacting the loan status. 

```python
#crosstab function to summarize table for each column

def crosstab(df,col):
    crosstab = pd.crosstab(df[col], df['loan_status'],margins=True)
    crosstab['Probability_Charged Off'] = round((crosstab['Charged Off']/crosstab['All']),3)
    crosstab = crosstab[0:-1]
    return crosstab

# Getting the Probability of charge off
def bivariate_probability(df,col,stacked= True):
    # get dataframe from crosstab function
    plotCrosstab = crosstab(df,col)  
    linePlot = plotCrosstab[['Probability_Charged Off']]      
    barPlot =  plotCrosstab.iloc[:,0:2]
    ax = linePlot.plot(figsize=(20,8), marker='o',color = 'b')
    ax2 = barPlot.plot(kind='bar',ax = ax,rot=1,secondary_y=True,stacked=stacked)
    ax.set_title(df[col].name.title()+' vs Probability Charge Off',fontsize=20,weight="bold")
    ax.set_xlabel(df[col].name.title(),fontsize=14)
    ax.set_ylabel('Probability of Charged off',color = 'b',fontsize=14)
    ax2.set_ylabel('Number of Applicants',color = 'g',fontsize=14)
    plt.show()
```

```python
#Checking for Purpose of Loan vs Probability for charge off

purpose=crosstab(df,'purpose')
display(purpose)
bivariate_probability(df,col='purpose',stacked=False)
```

Applicants who took loan for small business have the highest probability of charge off.
A guarantee or asset should be taken while approving the loan for this purpose.

```python
#check for annual income range vs probability chrge off

annual_inc_range=crosstab(df,'annual_inc_range')
display(annual_inc_range)
bivariate_prob(df=df,col='annual_inc_range')
```

Less annual income is indicating towards more defaulting in paying back ,percentage being 19%. Bank should consider this factor.

```python
#Checking for interest rates

int_rate_range=crosstab(df,'int_rate_range')
display(int_rate_range)
bivariate_probability(df,col='int_rate_range')
```

As the interest rates are increasing so are the default rates,highest being at 15%

```python
#Employment lenght and self employed check
emp_length=crosstab(df,'emp_length')
display(emp_length)
bivariate_probability(df,col='emp_length')
```

Applicants who are self employed or employment<1 year are more probable
to default.
